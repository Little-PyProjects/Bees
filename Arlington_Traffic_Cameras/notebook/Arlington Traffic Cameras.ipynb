```python
import pandas as pd
import folium
```

```python
df = pd.read_json('../data/Traffic+Cameras.json')
df.head(2)
```

#### This looks pretty good the Lat/Longs are already seperated and in decimal form. Let's get the centeroid and center the map on it

```python
lat = df['Latitude'].median()
long= df['Longitude'].median()
print(lat,long)
```

```python
m = folium.Map(location=[lat, long], zoom_start=13)
# m
```

```python
df.apply(lambda row:folium.Marker(location=[row["Latitude"], row["Longitude"]]).add_to(m), axis=1)
df.head(2)
```

#### The generic blue markers are pretty annoying. And I might as well add a popup.


```python
df.apply(lambda row:folium.Marker(location=[row["Latitude"], row["Longitude"]],
        popup=row['Camera Encoder'],icon=folium.Icon(color='red', icon="video-camera", prefix='fa'))
        .add_to(m), axis=1)
```
